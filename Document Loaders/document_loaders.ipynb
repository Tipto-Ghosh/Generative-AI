### PyPDFLoader
PyPDFLoader is a document loader in LangChain used to load content from PDF files and
convert **each page into a document object**.<br>

- It reads page by page and each page will be an object of document class.
- It will return a list where each element of the list is an object of document
- Limitation: It uses PyPDF library under the hood -- not great with scanned PDFs or 
complex layouts.

```python
from langchain_community.document_loaders  import PyPDFLoader
```

```python
pdf_file_path = Path("Document Loaders\dl-curriculum.pdf")
pdf_loader = PyPDFLoader(pdf_file_path)
```

```python
pdf_loader
```

```python
pdf_loader.file_path
```

```python
pdf_doc_object = pdf_loader.load()
```

```python
len(pdf_doc_object)
```

```python
print(pdf_doc_object[0].page_content)
```
